# GOT Wikipedia Dataset

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://en.wikipedia.org/wiki/List_of_Game_of_Thrones_episodes'
```

```python
page=requests.get(url)
```

```python
page
```

```python
soup=BeautifulSoup(page.text,'html.parser')
```

```python
print(soup.prettify)
```

```python
episodes_table=soup.find_all('table',class_='wikitable plainrowheaders wikiepisodetable')
```

```python
for table in episodes_table:
    rows=table.find_all('tr',class_='vevent')[0:]
    for items in rows:
        Season_number=items.find_all('td')[0].get_text()
        Title=items.find_all('td')[1].get_text()
        Director=items.find_all('td')[2].get_text()
        print(Season_number,Title,Director)
```

```python
episodes=[]

for table in episodes_table:
    headers=[]
    for header in table.find('tr').find_all('th'):
        headers.append(header.text)
    for rowss in table.find_all('tr')[1:]:
        values=[]
        for col in rowss.find_all(['th','td']):
            values.append(col.text)
        episodes_dict={headers[i] : values[i] for i in range(len(values))}
        episodes.append(episodes_dict)
```

```python
df=pd.DataFrame(episodes)
```
