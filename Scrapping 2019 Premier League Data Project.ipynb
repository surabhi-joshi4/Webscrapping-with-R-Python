```python
import requests
```

```python
url='https://www.skysports.com/premier-league-table/2019'
```

```python
page=requests.get(url)
```

```python
print(page.status_code)
```

```python
print(page)
```

```python
#print(page.text)
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(page.text,'html.parser')
```

```python
#Beautiful Soup is a Python package for parsing HTML and XML documents. It creates a parse tree for parsed pages that can be used to extract data from HTML, which is useful for web scraping.
#Prettify() function in BeautifulSoup will enable us to view how the tags are nested in the document.
```

```python
print(soup.prettify)
```

```python
#find is use to search the tag 
print(soup.find('a'))
```

```python
#find_all is use to search all the tags in webpage of which we specified in parameter 
print(soup.find_all('a'))
```

```python
#find the tag with the specify index
soup.find_all('a')[1]
```

```python
league=soup.find('table',class_='standing-table__table')
```

```python
print(league)
```

```python
league_table=league.find_all('tbody')
```

```python
print(league_table)
```

```python
#text.strip() function will return text contained within this tag and strip any kind of extra spaces, '\n','\t' from the text string object.
```

```python
for league_teams in league_table:
    rows=league_teams.find_all('tr')
    for row in rows:
        team_names=row.find('td',class_='standing-table__cell standing-table__cell--name').text.strip()
        team_points=row.find_all('td',class_='standing-table__cell')[9].text.strip() #9=data_index_id
        print(team_names,team_points)        
```

```python
import pandas as pd
```

```python
league_2011=[]
for league_teams in league_table:
    rows=league_teams.find_all('tr')
    for row in rows:
        team_names=row.find('td',class_='standing-table__cell standing-table__cell--name').text.strip()
        team_points=row.find_all('td',class_='standing-table__cell')[9].text.strip() #9=data_index
        league_dict={'name':team_names,
                    'pts':team_points}
        league_2011.append(league_dict)
```

```python
df=pd.DataFrame(league_2011)
```
